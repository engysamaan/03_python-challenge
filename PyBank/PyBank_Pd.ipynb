```python
import pandas as pd
```

```python
# Save path to data set in a variable
budget_data = "Resources/budget_data.csv"
```

```python
# Use Pandas to read data
budget_data = pd.read_csv(budget_data)
budget_data.head()
```

```python
total_months=budget_data["Date"].count()
total_months
```

```python
total_amt=budget_data["Profit/Losses"].sum()
total_amt
```

```python
budget_data["prev_amt"] = budget_data["Profit/Losses"].shift(1)
budget_data["average_change"]= budget_data["Profit/Losses"] - budget_data["prev_amt"]
#budget_data["average_change"] = budget_data["average_change"].abs()
```

```python
budget_data.sort_values("Profit/Losses",ascending=False)
```

```python
greatest_increase=budget_data["average_change"].max()
greatest_increase
```
