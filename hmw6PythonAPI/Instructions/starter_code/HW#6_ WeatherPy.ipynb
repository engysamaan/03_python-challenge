## Generate Cities List

```python
#1- giting the lat and logs of the nerest cities from the citypy

lat = np.random.uniform(low=-90.000, high=90.000, size=1500)
lng = np.random.uniform(low=-180.000, high=180.000, size=1500)

lat[0], lng[0]
```

```python
#1.2 loop to get the lats and lngs of the number of cities we want
lat_lng = []

for i in range(len(lat)):
    i=lat[i],lng[i]
    print(i)
    lat_lng.append(i)
#lat_lng[0] , lat_lng[1]
```

```python
#testing
citipy.nearest_city(lat[0],lng[0]).city_name
citipy.nearest_city(23.812511410155423, 175.72332981054797).country_code
```

```python
#1.3 get the cities data of the above lat_lng list

cities = []

for loc in lat_lng:
    print(f"This is city nu # " + str(len(cities)+1))
    print (loc)
    lat = loc[0]
    lng = loc[1]

    city = citipy.nearest_city(lat,lng).city_name , citipy.nearest_city(lat,lng).country_code
    
    print(city)
    print("*--------------------------------*")
    cities.append(city)


#print(cities)   
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#2 read the url of each city by json
base_url = "http://api.openweathermap.org/data/2.5/weather?"
#api_keys = "a996ea448805ee4d39861e051d472592"
units="Imperial"

query_url = f"{base_url}appid={api_key}&units={units}&q="

city_weather = []

for city in cities:
    city_url = req.get(query_url + city[0]).json()
    #print(city_url)
    try:
        # Parse out the max temp, humidity, and cloudiness
        city_name = city_url['name']
        city_code = city_url['sys']['country']
        city_lat = city_url["coord"]["lat"]
        city_lng = city_url["coord"]["lon"]
        city_max_temp = city_url["main"]["temp_max"]
        city_humidity = city_url["main"]["humidity"]
        city_clouds = city_url["clouds"]["all"]
        city_wind = city_url["wind"]["speed"]
        city_date = city_url["dt"]
        
        
        city_url =  {"City"     : city_name,
                    "Country"   : city_code,
                    "Lat"       : city_lat,
                    "Lng"       : city_lng,
                    "Max Temp"  : city_max_temp,
                    "Humidity"  : city_humidity,
                    "Cloudiness": city_clouds,
                    "Wind Speed": city_wind,
                    "Date"      : city_date}
        
        city_weather.append(city_url)
    
    except:
        print("City not found...")
        pass
 
    
    

print(len(city_weather))
```
